```python
#Importing neccesary libraries

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline
```

## **Part 1: Using figures and axes (Object oriented method)**

### 1.1 Line plot

```python
#creating figure object
fig=plt.figure()

#printing the type of fig
print(type(fig))

#creating getting current axes from the figure
ax=fig.gca()

#printing the type of ax
print(type(ax))

# Let's plot a line
# Create two numpy arrays x and y
x=np.arange(0,10,1)
y=3*x

#plot is a method in axes class used to plot our data
#we use a keyword 'label' to label our plot.
ax.plot(x,y, label='a function y=5x')

#Labelling x and y axis.
ax.set_xlabel('x value')
ax.set_ylabel('y value')

#setting a title to our figure
ax.set_title('Plotting a function y=5x')

#call plot.show() function to display our plot.
plt.show()
```

### 1.2 Sine and cosine using line plot

```python
#lets plot sine and cosine curves

#Using np.linspace to create a numpy array with 100 points between 0 and 8
x=np.linspace(0,8,100)

y_sin=np.sin(x)
y_cos=np.cos(x)

#we can plot two plots in the same figure

#creating figure object
fig=plt.figure()

#getting the current axes
ax=fig.gca()

#lets plot sine curve first, by calling plot() method
ax.plot(x, y_sin, label='sin curve')

#plotting cos curve
ax.plot(x, y_cos, label='cos curve')

ax.set_xlabel('x value')
ax.set_ylabel('y value')

#call legend() method on axes object to display legend on the axes
ax.legend()

#you can specify a keyword 'fontsize' to change fontsize
ax.set_title('Both sine and cosine curves in the same axes', fontsize=15)

plt.show()
```

### 1.3 Scatter plot

```python
#let's plot a scatter plot

#Generating 500 random numbers from a standard normal distribution
x=np.random.randn(500)
y=np.random.randn(500)

#you can specify the keyword 'figsize' in figure() to change the figure size
fig=plt.figure(figsize=(7,7))

ax=fig.gca()

#Use scatter method to plot a scatter plot
#You can specify keywords: 'c' or 'color' to specify color, 
#'s' to specify the marker size, 'marker' to speficy marker type
ax.scatter(x,y, label='scatter plot', color='g', s=10, marker= 'o')

ax.set_xlabel('x')
ax.set_ylabel('y')

ax.legend()

ax.set_title('Plotting scatter plot')

plt.show()
```

## **Part 2:Using pyplot**

![Anatomy of a plot](https://i.ibb.co/p0kx3c1/anatomy-plot.png)
     

For every method(like plot, set_xlabel, set_title, etc.) on axes class, there's a corresponding function while using pyplot.<br/>
Below you can see the abridged version of plot function.

```python
#In matplotlib's pyplot.py

#This is for demonstrating how functions in pyplot works under the hood.

def plot(*args, **kwargs):
    
    """An abridged version of pyplot.plot()."""
    ax=plt.gca()
    return ax.plot(*args, **kwargs)

def gca(**kwargs):
    
    #gets the current figure
    fig=plt.gcf()
    
    #gets the current axes and returns it.
    return fig.gca(**kwargs)
```

We can see that all the functions under pyplot just wrap the object oriented interface and provides a convenient way to work with the current figure and axes

### 2.1 Line plot

```python
# Let's plot a line
# Create two numpy arrays x and y

x=np.arange(0,10,1)
y=3*x

#note that we have imported matplotlib.pyplot as plt

#plot function creates figure and axes objects for you
plt.plot(x,y, label='a function y=5x')

#Labelling x and y axis.
plt.xlabel('x value')
plt.ylabel('y value')

#setting a title to our figure
plt.title('Plotting a function y=5x')

#call plot.show() function to display our plot.
plt.show()
```

Similarly, the functions xlabel, title in pyplot work as below.

```python
#In matplotlib's pyplot.py

#plt.xlabel

def xlabel(s,*args, **kwargs):
    #gets the current axes
    ax=plt.gca()
    
    #calls set_xlabel method on the current axes object
    ax.set_xlabel(s,*args, **kwargs)

#Similarly, plt.title

def title(s,*args, **kwargs):
    #gets the current axes
    ax=plt.gca()
    
    #calls set_title method on the current axes object
    ax.set_title(s,*args, **kwargs)
```

## **Part 3: Subplots**

<p style="text-align: center;">Anatomy of Figure and Axes</p>

![Anatomy of figure and axes](https://i.ibb.co/cxWrsjq/anatomy-subplots-1.jpg)

```python
#create figure and axes objects by calling plt.subplots()

#we are only creating one axes on the figure
fig, ax = plt.subplots()

#let's print the types of fig and ax returned by plt.subplots()
print(f'The type of fig is {type(fig)}')
print(f'The type of ax is {type(ax)}')

#lets plot y=x**2 and y=x**3 on the same plot

#generating 200 points between -2 and 2
x=np.linspace(-2,2,200)

y_2=x**2
y_3=x**3

ax.plot(x,y_2, label='y=x**2')
ax.plot(x,y_3, label='y=x**3')


#setting x and y labels using ax.set_xlabel() and ax.set_ylabel()
ax.set_xlabel('x')
ax.set_ylabel('y')

#to show the legend
ax.legend()
#you can set title using ax.set_title()

ax.set_title('Plotting y=x\u00b2 and y=x\u00b3')

plt.show()
```

```python
#Creating multiple axes on the figure
fig, ax = plt.subplots(nrows=2,ncols=2)

#let's print the types of fig and ax returned by plt.subplots()
print(f'The type of fig is {type(fig)}')
print(f'The type of ax is {type(ax)} \n')

print(ax)
```

* We can see that when we create multiple axes on the same figure, plt.subplots returns the current figure object and a numpy array of axes object, each corresponding to each axes(subplots).
* Here, the plots are empty because we have not plotted anything on the axes.


```python
#lets plot the same y=x**2 and y=x**3 on different plots on the same figure.

fig, ax=plt.subplots(nrows=1,ncols=2, figsize=(10,5))

#we access the first axes using ax[0]
#plot y=x**2 on the first axes
ax[0].plot(x,y_2, label='y=x**2')

#each axes has it's own labels, title, legend.
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[0].legend()
ax[0].set_title('Plot of Y=X**2')


#plot y=x**3 on the second axes
ax[1].plot(x,y_3, label='y=x**3', color='orange')

ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

ax[1].legend()
ax[1].set_title('Plot of Y=X**3')

#We can title the whole figure using fig.suptitle()
fig.suptitle('Plots of both y=x\u00b2 and y=x\u00b3 on different axes', fontsize=15)

#you can save your figures/plots using plt.savefig()
plt.savefig('plot_x_x2.jpg', dpi=150)

plt.show()
```

```python
#Let's 5 y=x**k plots from k=1 to 5 using subplots

x=np.linspace(-1,1,100)

fig,ax=plt.subplots(ncols=3, nrows=2, figsize=(10,6))

# since plt.subplots return a numpy array of axes objects, we can flatten it using .ravel()
ax=ax.ravel()

#lets color each plot differently.
colors_list=['r','g','b','yellow','black']


for i in range(5):
    ax[i].plot(x, x**(i+1), color=colors_list[i])
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    ax[i].set_title(f'y=x**{i+1}')

fig.suptitle('Plots of y=x**k, with k from 1 to 5', fontsize=15)
plt.show()
```

We can see that last axes is left empty but axis are still present. (Do not confuse axes with axis) /n We can prevent this by using plt.subplot instead.

## Subplot

```python
fig=plt.figure(figsize=(10,6))


for i in range(5):
    ax=plt.subplot(230 +i+1) # 2 rows 3 columns and ith index
    ax.plot(x, x**(i+1),color=colors_list[i])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title(f'y=x**{i+1}')

fig.suptitle('Plots of y=x**k, with k from 1 to 5', fontsize=15)
plt.show()
```

The main difference between plt.subplot and plt.subplots is that plt.subplots creates all the axes first and draws on the, while plt.subplot creates each one on the go.
